```python
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
# Load the English subset of the wiki40b dataset
ds, ds_info = tfds.load('wiki40b/en', with_info=True, split='train', download=True)

# Print dataset information to understand its structure
print(ds_info.description)

# Display example entries from the dataset
for example in ds.take(5):  # Adjust the number for more examples
    print(example['text'].numpy().decode('utf-8'))  # Decode from bytes to string
```

```python
keywords = ['engineering', 'machine learning', 'data science', 'analysis', 'biological science']
```

```python
def filter_articles(text):
    text = text.numpy().decode('utf-8').lower()
    # Check if any keyword is present in the text
    return any(k in text for k in keywords)

# Apply filter on each article
filtered_ds = ds.filter(lambda x: tf.py_function(func=filter_articles, inp=[x['text']], Tout=tf.bool))
```
